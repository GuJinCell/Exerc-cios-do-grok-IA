````python
def inverter_por_tamanho(texto, tamanho):
    """
    Recebe um texto e cria uma lista com as palavras invertidas que forem maiores que o tamanho dado

    Args:
        texto(str): Texto com palavras separadas por espaço
        tamanho(int): Parâmetro de tamanho mínimo para inverter uma palavra
    
    Returns:
        list: Lista das palavras que foram invertidas por serem do tamanho adequado

    Raises:
        ValueError: Se o tamanho ou o texto forem inválidos
    """
    if not isinstance(texto,str) or tamanho < 0:
        raise ValueError
    if not texto.strip():
        return []
    palavras = texto.split()
    def processar(palavras):
        if not palavras:
            return []
        primeiro = palavras[0]
        resto = palavras[1:]
        if len(primeiro) >= tamanho:
            return [primeiro[::-1]] + processar(resto)
        return processar(resto)
    return processar(palavras)
    


texto = "casa sol lua python teste"
tamanho = 4
print(inverter_por_tamanho(texto,tamanho))
````

### Questão 2: Agrupamento de Linhas em Arquivo por Palavra-Chave
##### Escreva uma função agrupar_por_palavra(arquivo_entrada, arquivo_saida, palavra) que leia um arquivo de texto, agrupe as linhas que contêm a palavra-chave (ignorando maiúsculas/minúsculas) em um novo arquivo, e adicione no final a contagem de linhas agrupadas. Use try-except para erros de arquivo.
- Entrada:
arquivo_entrada.txt: 

Casa sol

Lua PYTHON

teste casa

sol lua


palavra = "casa"

- Saída em arquivo_saida.txt:

Casa sol

teste casa

Total de linhas com 'casa': 2

- Restrições: Levante FileNotFoundError se o arquivo de entrada não existir. Ignore case na busca pela palavra.

````python
import os

def agrupar_por_palavra(entrada, saida, palavra):
    """
    Recebe um arquivo de entrada com frases e uma palavra como parâmetro
    para as frases que vão ser registradas no arquivo saída

    Args:
        entrada(str): Arquivo lido inicialmente
        saida(str): Arquivo de saída somente com frases contendo a palavra
        palavra(str): Palavra parâmetro para o arquivo saída
    
    Raises:
        FileNotFoundError: Se o arquivo de entrada não existir.
        ValueError: Se palavra não for uma string não vazia.
    """
    contador = 0
    try:
        if not palavra or not isinstance(palavra,str):
            raise ValueError("Palavra é inválida")
        with open(entrada,"r", encoding="utf-8") as arq:
            with open(saida,"w",encoding="utf-8") as res:
                for i in arq.read().splitlines():
                    if palavra.lower() in i.lower():
                        res.write(i + "\n")
                        contador += 1
                res.write(f"\nTotal de linhas com '{palavra}': {contador}")
    except FileNotFoundError:
        raise FileNotFoundError("Arquivo não encontrado")
    

pasta = "C:/Users/Celso.O.C/PycharmProjects/progamation/Exercícios/Exerc-cios-do-grok-IA/arquivos_de_exercicios/arquivos_dias_31_40"
os.makedirs(pasta,exist_ok=True)
entrada = f"{pasta}/arquivo_ent_40.txt"
saida = f"{pasta}/result_ent_40.txt"
try:
    with open(entrada,"w",encoding="utf-8") as arquivo:
        arquivo.write("Casa sol\nLua PYTHON\nteste casa\nsol lua")
    agrupar_por_palavra(entrada,saida,"casa")
except FileNotFoundError as f:
    print(f)
````

### Questão 3: Herança em Classes para Produtos
##### Crie uma classe base Produto com atributos nome (string) e preco (float), e métodos:
__init__ para inicializar os atributos.

__str__ para retornar "Produto: nome, Preço: R$preco".

aplicar_desconto(desconto) para reduzir o preço (em percentual).

##### Crie uma classe derivada ProdutoPerecivel que herda de Produto, com atributo adicional validade (string) e:
__init__ que chama o construtor da classe base e inicializa validade.

-- Sobrescreva __str__ para incluir a validade (ex.: "Produto: nome, Preço: R$preco, Validade: validade").

-- Sobrescreva aplicar_desconto para aplicar um desconto extra de 10% se a validade for "hoje".

- Exemplo:
python

p1 = Produto("Arroz", 20.0)

p2 = ProdutoPerecivel("Leite", 10.0, "hoje")

print(p1)  # Produto: Arroz, Preço: R$20.0

print(p2)  # Produto: Leite, Preço: R$10.0, Validade: hoje

p1.aplicar_desconto(10)

p2.aplicar_desconto(10)

print(p1)  # Produto: Arroz, Preço: R$18.0

print(p2)  # Produto: Leite, Preço: R$8.1  (10% + 10% extra = 19% de desconto)

````python
class Produto:
    """Classe base para produtos com nome e preço."""
    def __init__(self, nome, preco):
        if not isinstance(nome,str) or not nome:
            raise ValueError("Nome inválido")
        if not isinstance(preco,(int,float)) or preco < 0:
            raise ValueError("Preço inválido")
        self.nome = nome
        self.preco = preco

    def __str__(self):
        """Retorna a representação do produto."""
        return f"Produto: {self.nome}, Preço: R${self.preco:.2f}"
    
    def aplicar_desconto(self, desconto):
        """Aplica um desconto percentual ao preço."""
        if not isinstance(desconto, (int, float)) or desconto < 0:
            raise ValueError("Desconto deve ser um número não negativo")
        self.preco = self.preco * (1 - (desconto/100))

class ProdutoPerecivel(Produto):
    """Classe para produtos perecíveis com validade."""
    def __init__(self, nome, preco, validade):
        super().__init__(nome, preco)
        if not isinstance(validade,str):
            raise ValueError("Validade deve ser uma string")
        self.validade = validade
    
    def __str__(self):
        """Retorna a representação do produto perecível."""
        return f"Produto: {self.nome}, Preço: R${self.preco:.2f}, Validade: {self.validade}"
    
    def aplicar_desconto(self, desconto):
        """Aplica desconto com extra de 10% se validade for 'hoje'."""
        super().aplicar_desconto(desconto)
        if self.validade.lower() == "hoje":
            super().aplicar_desconto(10)    



p1 = Produto("Arroz", 20.0)

p2 = ProdutoPerecivel("Leite", 10.0, "hoje")

print(p1)  # Produto: Arroz, Preço: R$20.0

print(p2)  # Produto: Leite, Preço: R$10.0, Validade: hoje

p1.aplicar_desconto(10)

p2.aplicar_desconto(10)

print(p1)  # Produto: Arroz, Preço: R$18.0

print(p2)  # Produto: Leite, Preço: R$8.1  (90% - 10% extra = 19% de desconto)
````

### Questão 4: Serialização de Dicionário em JSON
##### Escreva uma função salvar_dicionario_json(dicionario, arquivo_saida) que receba um dicionário e salve seus dados em um arquivo JSON com formatação legível (indentação). Use try-except para lidar com erros de escrita. A função deve retornar True se a operação for bem-sucedida, False caso contrário.
- Entrada:
dicionario = {"nome": "Ana", "idades": [20, 25], "ativo": True}

arquivo_saida = "dados.json"

- Saída em dados.json:
```json

{
    "nome": "Ana",
    "idades": [
        20,
        25
    ],
    "ativo": true
}
```
- Restrições: Use a biblioteca json. Trate erros como permissões de escrita.

````python
import json

def salvar_dicionario_json(dicionario, saida):
    """Salva um dicionário em um arquivo JSON com formatação legível.
    
    Args:
        dicionario (dict): Dicionário a ser serializado.
        arquivo_saida (str): Caminho do arquivo de saída.
    
    Returns:
        bool: True se a operação for bem-sucedida, False caso contrário.

    Raises:
        ValueError: Se o dicionário não for dict ou se a saída não for str
    """
    if not isinstance(dicionario,dict):
        raise ValueError("Dicionário inválido, deve ser dict")
    if not isinstance(saida,str) or not saida:
        raise ValueError("Saída inválida, deve ser string")
    try:
        with open(saida, "w", encoding="utf-8") as f:
            json.dump(dicionario, f, indent=4,ensure_ascii=False)
        return True
    except (IOError, TypeError):
        return False

dicionario = {"nome": "Ana", "idades": [20, 25], "ativo": True}
arquivo_saida = "dados_40.json"
print(salvar_dicionario_json(dicionario,arquivo_saida))
````

### Questão 5: Filtro Recursivo de Estrutura Aninhada
##### Escreva uma função recursiva filtrar_positivos_aninhados(lista) que receba uma lista possivelmente aninhada (contendo números ou sublistas) e retorne uma nova lista com apenas os números positivos, preservando a estrutura aninhada, sem usar loops.
- Entrada: lista = [1, -2, [3, -4, [5, -6]], 7]

- Saída: [1, [3, [5]], 7]
Remove números negativos, mantém a estrutura.

- Restrições: Não use loops (for ou while). Use isinstance para verificar sublistas.

````python
def filtrar_positivos_aninhados(lista):
    """Filtra números positivos de uma lista aninhada, preservando a estrutura.
    
    Args:
        lista (list): Lista possivelmente aninhada com números ou sublistas.
    
    Returns:
        list: Lista com números positivos, mantendo a estrutura aninhada.

    Raises:
        ValueError: Se o argumento não for uma lista.
    """
    if not isinstance(lista,list):
        raise ValueError("Lista inválida")
    if not lista:
        return []
    primeira = lista[0]
    resto = lista[1:]
    if isinstance(primeira,(int,float)):
        if primeira > 0:
            return [primeira] + filtrar_positivos_aninhados(resto)
        return filtrar_positivos_aninhados(resto)
    if isinstance(primeira,list):
        return [filtrar_positivos_aninhados(primeira)] + filtrar_positivos_aninhados(resto)
    return filtrar_positivos_aninhados(resto)
        


lista = [1, -2, [3, -4, [5, -6]], 7]
print(filtrar_positivos_aninhados(lista))
````
